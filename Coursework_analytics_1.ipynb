Для того, чтобы проверить, что вы правильно соединили все датасеты в основной (без `students.csv`), запустите аналогичный приведенному ниже запрос. Выдача должна совпадать. Имейте в виду, что названия некоторых столбцов были переименованы.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from datetime import datetime as dt
```

```python
# Запишем данные из csv файлов в датафреймы и проведём предварительную обработку(переименуем столбцы, выбросим лишнее)

# Курсы
# Столбец id переименовываем в course_id для однозначности при дальлнейшем объединении датафреймов.
Courses = pd.read_csv('courses.csv') \
    .rename(columns={'id': 'course_id'}) \
    .drop(['Unnamed: 0'], axis=1)

# Студенты
Students = pd.read_csv('students.csv',parse_dates=[-1])

# Содержание курсов
Course_contents = pd.read_csv('course_contents.csv')

# Прогресс
# Переименовываем столбец id в progress_id
Progresses = pd.read_csv('progresses.csv') \
    .rename(columns={'id': 'progress_id'})

Progress_phases = pd.read_csv('progress_phases.csv')
```

```python
# Перед тем как объединять датафреймы в сводные для последующей работы - посмотрим информацию по ним.
# 1. выведем информацию об отсутствующих данных по всем датафреймам.
print(Courses.isna().sum(), '\n')
print(Students.isna().sum(), '\n')
print(Course_contents.isna().sum(), '\n')
print(Progresses.isna().sum(), '\n')
print(Progress_phases.isna().sum(), '\n')

# Видим, что отсутствующие данные есть всего в двух датафреймах. 
# Нет данных о городах и днях рождения большого количества студентов.
# Также есть пропущенные данные в столбце finish_date датафрейма progress_phases.
# Вероятно это незаконченные домашние работы, курсовые и т.д.
# В целом данные достаточно полные. Это обнадёживает.
```

```python
#объединяем таблицы, связаны с курсом, по общему идентификатору
cors = pd.merge(Courses, Course_contents, on = 'course_id')
#объединяем таблицы, связаны с прогрессом, по общему идентификатору
prog = pd.merge(Progresses,Progress_phases, on = 'progress_id')

#объединяем все таблицы, кроме "students.csv", по общим идентификаторам
main = pd.merge(cors, prog, on = ['course_id','module_number','lesson_number'])
main.head()
```

#  1.


```python
# Общее количество курсов в датасете
print(len(pd.unique(main['title'])));
```

```python
# Количество модулей на каждом курсе
print('Количество модулей на каждом курсе')
main.groupby('title').nunique().module_title.sort_values(ascending=False)
```

```python
# Количество уроков в каждом модуле на каждом курсе
les_by_mod_by_course = main.groupby(['title','module_number'])['lesson_title'].nunique().unstack(level=0)
les_by_mod_by_course
```

```python
# Медианное количество уроков в модуле на каждом курсе
# Просто применяю к предыдущему датафрейму функцию median по столбцам.
les_by_mod_by_course.apply('median',axis=0)
```

```python
# Количество учеников на каждом курсе
students_count_by_course = main.groupby('title').nunique().student_id.sort_values(ascending=False)
students_count_by_course
```

```python
# Минимальный, максимальный, средний и медианный возраст студентов.
# Предполагаю, что данные нам нужны только из общего датасета.
# Возраст считаем на последнюю дату по столбцам start_date и finish_date. Считать возраст на текущий день некорректно.
# tz_localize(None) убирает данные о таймзоне. Они не сильно нужны, но мешают расчёту возраста, поскольку дни рождения
# в датафрейме указаны без таймзоны.

#cоздаем копию датафрейма
age_ = Students.copy()

#удаляем ненужныем столбцы
age_.drop(['id_' , 'city'], axis=1, inplace=True)

#соеденяем 2 датафрейма по id, которые имеют разные название в разных датафреймах
age_ = pd.merge(age_, main, left_on = 'id', right_on = 'student_id')

#оставляем только нужные столбцы
age = age_[['id','birthday','start_date']]

#переводим строку в дату, стандартизируем дату
age['start_date'] = pd.to_datetime(age.start_date).dt.tz_localize(None)

#приводим дату к удобнему для нас типу
age['start_date'] = age['start_date'].apply(lambda x: x.strftime('%Y-%m-%d')) 
age['start_date'] = pd.to_datetime(age.start_date)

#находим разницу между датой рождения и первым выполненным заданием  в годах
age['Years'] = (age.start_date - age.birthday).astype('timedelta64[Y]')

#берем только те даты, которые +- похожи на реальные
age = age.loc[(age.birthday >='1950/1/1') & (age.birthday <='2003/12/31'), :]
age.head()
```

```python
# Минимальный, максимальный, средний, медианный возраст студентов
print(f'Минимальный возраст   {(age.Years.min())}\
       \nМаксимальный          {(age.Years.max())}\
       \nСредний               {round((age.Years.mean()),2)}\
       \nМедианный             {(age.Years.median())}')
```

```python
# Данные о возрасте студентов по каждому из курсов
# Группируем по курсам и сразу считаем все 4 величины
ages_by_course = age2.groupby('title').agg({'Years':['max','min','mean','median']})
# Убираем иерархический индекс у столбцов.
ages_by_course.columns = ages_by_course.columns.droplevel(0)
# Добавляем префикс age к именам столбцов, для лучшей читаемости.
ages_by_course.rename(columns={'max':'age_max', 'min':'age_min', 'mean':'age_mean', 'median':'age_median'}, inplace=True)
print('Данные о возрасте студентов по каждому курсу:')
ages_by_course.sort_values(by='age_median', ascending=False)
```

```python
plt.figure(figsize=(10,5))
ax = plt.bar(students_count_by_course.index, students_count_by_course)
# Подписи столбцов поворачиваем, выравниваем по правому краю и привязываем к столбцу 
plt.title('Количество студентов по курсам', fontsize=10)
plt.xticks(rotation=30, ha='right', fontsize=10)
# plt.yticks(fontsize=10)
plt.ylabel(ylabel='Количество студентов', fontsize=12)
plt.show()
```

```python
# Отдельно для этого графика создаём датафрейм с дополнительной колонкой field и сразу отсортируем по убыванию.
df1_3 = main.groupby('title').agg({'student_id':'nunique', 'field':'first'}).\
                sort_values(by='student_id', ascending=False)

# Создадим словарь цветов для каждой сферы.Он нам пригодится для легенды.
# Можно и вручную было цвета прописать, но через формирование цветовой палитры можно свободно менять их.
mapper = sns.color_palette('deep',df1_3.field.unique().size)
color_dict = {}
for num, field in enumerate(df1_3.field.unique()):
    color_dict[field] = mapper[num]

# А теперь сформируем палитру для графика:
my_palette = []
for element in df1_3.field:
    my_palette.append(color_dict[element])

plt.figure(figsize=(20,10))
# По-поводу альфы не совсем понятно в задании. Альфа 0.9 означает прозрачность 0.1 и наоборот соответственно.
# При альфе 0.1 почти ничего не будет видно, поэтому я предположил, что имелась в виду альфа 0.9
students_hbar = sns.barplot(x='student_id', y=df1_3.index, data=df1_3, palette=my_palette, alpha=0.9)

# Отрисуем вертикальную линию по медианному количеству студентов.
fl = plt.axvline(df1_3.student_id.median(), ls='--', lw=3, color='purple', alpha=0.9, label='Медианное количество студентов')
# Добавим для неё отдельный xtick
xt = students_hbar.get_xticks() 
xt = np.append(xt,df1_3.student_id.median())
students_hbar.set_xticks(xt)

# И отдельную легенду для этой линии.
first_legend = plt.legend(handles=[fl], loc=7, fontsize = 20, shadow=True)
# Теперь добавим эту легенду в текущий axes
plt.gca().add_artist(first_legend)

# Нарисуем вручную прямоугольники для легенды и присвоим им цвета из нашего словаря цветов.
handles = [plt.Rectangle((0,0),1,1, color=color_dict[key], alpha=.9) for key in color_dict.keys()]
# Отрисовываем вторую легенду.
plt.legend(handles, color_dict.keys(), fontsize=20, title='Field', title_fontsize=25, shadow=True)

# Настроим подписи графика, осей, меток.
plt.title('Количество студентов по курсам', fontsize=25)
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.xlabel('Количество студентов', fontsize=20)
plt.ylabel(None)

# Убираем рамку и отрисовываем вертикальные линии сетки, для лучшей читаемости диаграммы.
plt.box(False)
plt.grid(axis='x', alpha=0.3, ls='--', lw=2)
plt.show()
```

```python
# В этом и последующих блоках при формировании датафрейма для работы названия им даю по номеру блока.

# Сначала сделал такой вариант, группировка, реформат и две сортировки работают достаточно быстро.
# 1. Выбираем только те строки которые относятся к домашним работам. Группируем по курсу и по айди студента,
# выбираем самую раннюю дату начала домашней работы и сбрасываем индекс. 
df2_1 = main[main['is_homework'] == True].groupby(['title', 'student_id']).start_date.min().reset_index()
# 2. Создаем столбец с отформатированными датами. Нам нужен только месяц и год.
df2_1['year_month'] = pd.to_datetime(df2_1.start_date).apply(lambda x: x.strftime('%Y-%m'))
# 3. Группируем по курсу и дате и считаем количество студентов. Unstack здесь нужен для того, чтобы курсов была
# одинаковая шкала дат.
df2_1 = df2_1.groupby(['title','year_month']).student_id.count().unstack(level=0).fillna(0)
# Здесь всё работает достаточно хорошо, но я понял, что такой вариант не совсем правильный. Студент может начать выполнять
# домашние работы в модуле не по порядку. Поэтому я собрал датафрейм по-другому.


# 1. Собираем датафрейм с номерами в которых впервые появляется домашнее задание в модуле. 
# Сначала отфильтровываем только те строки, которые относятся к домашним работам.
# затем группируем по курсам и модулям и выбираем уроки с минимальным номером. Сбрасываем индекс.
# снова группируем уже только по курсам и выбираем первые строки. В итоге получаем список номеров модулей и уроков,
# где впервые появляется домашняя работа.
first_lessons = main.loc[main['is_homework'] == True].groupby(['title','module_number']).agg({'lesson_number':'min'})\
                .reset_index().groupby('title').first().reset_index()
# 2. Теперь с помощью merge способом left из основного датафрейма оставим только нужные нам строки.
df2_1_best = first_lessons.merge(main[main['is_homework'] == True],\
                                 on=['title','module_number','lesson_number'], how='left')

# 3. Создаем столбец с отформатированными датами. Нам нужен только месяц и год.
df2_1_best['year_month']=pd.to_datetime(df2_1_best.start_date).apply(lambda x: x.strftime('%Y-%m'))
# 4. Также группируем по курсу и дате и считаем количчество студентов. Опять же unstack здесь нужен в основном для того,
# чтобы для всех курсов была одинаковая временная шкала.
df2_1_best = df2_1_best.groupby(['title','year_month']).student_id.nunique().unstack(level=0).fillna(0)

# Вот так выглядит итоговый датафрейм.
df2_1_best
```
